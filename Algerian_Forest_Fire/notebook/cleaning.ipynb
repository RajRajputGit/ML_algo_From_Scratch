### fix the data type of columns

```python
df[['month','day','year','Temperature','RH','Ws','Region']] =df[['month','day','year','Temperature','RH','Ws','Region']].astype('int64')
```

```python
df['Rain']  = df['Rain'].astype('float64')
df['FFMC']  = df['FFMC'].astype('float64')
df['DMC']  = df['DMC'].astype('float64')
df['ISI']  = df['ISI'].astype('float64')
df['BUI']  = df['BUI'].astype('float64')
```

```python
df.head()
```

### THe two columns contains bad values therfore we have to fix those then we can change their data type
### columns with bad values are 'DC' and 'FWI'

```python
# Step 1: Strip whitespace within the string (e.g., '14.6 9' → '14.69')
df['DC'] = df['DC'].astype(str).str.replace(r'\s+', '', regex=True)

# Step 2: Convert to float, coercing any remaining bad values to NaN
df['DC'] = pd.to_numeric(df['DC'], errors='coerce')

# Step 3: Inspect how many NaNs were introduced
print(df['DC'].isna().sum())
```

```python
col = 'FWI'
# Step 1: Strip whitespace
df[col] = df[col].astype(str).str.strip()

# Step 2: Replace known junk strings with NaN
df[col] = df[col].replace({'fire': np.nan, 'nan': np.nan, '': np.nan})

# Step 3: Safe numeric conversion
df[col] = pd.to_numeric(df[col], errors='coerce')

# Step 4: Check NaN count introduced
print(df[col].isna().sum())

df.dropna(subset=col).reset_index(drop=True, inplace=True)
```
